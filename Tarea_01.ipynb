Problema 8.- Escribir un programa que pida al usuario un número entero y muestre en pantalla un tríangulo rectángulo el que se muestra.

```
def triangulo (n):
  """
  for i in range(1,n+1)
  "*"*i
  []
  "\n".join()
  """
  return "\n".join(["*"*i for i in range(1,n+1)])

numero=int(input("Indica el numero: "))
print(triangulo(numero))
```

Problema 9.- Escribir una función que reciba un número entero positivo y devuelva su factorial.

```
n= int (input("ingrese el factorial que desea realizar:" ))
def factorial(n):
  if n==0:
    return 1
  else:
    return n * factorial(n-1)
print("el factorial de:",n,"es:",factorial(n))
```
